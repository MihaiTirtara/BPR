```python
documents = []

import os

for dirpath, dirnames, files in os.walk(r"D:\rvl-cdip\images\imagesa\a\a"):
    for file in files:
        if file.endswith(".jpg"):
         print(os.path.join(dirpath, file))
         img = cv2.imread(os.path.join(dirpath, file))
         documents.append(pytesseract.image_to_string(img, config=custom_config))
        
```

```python
import re

counter=0

fp0 = open(r"D:\rvl-cdip\labels\test.txt", 'r')
fp1 = open(r"D:\rvl-cdip\labels\train.txt", 'r')
fp2 = open(r"D:\rvl-cdip\labels\val.txt", 'r')

text0 = fp0.read()
text1 = fp1.read()
text2 = fp2.read()

result = list()

for dirpath, dirnames, files in os.walk(r"D:\rvl-cdip\images\imagesa\a\a"):
    for file in files:
        if file.endswith(".jpg"):
            file_ext = file.replace("+", "\+")
            rez0 = re.findall("/" + file_ext + " (\d{2}|\d)", text0)
            rez1 = re.findall("/" + file_ext + " (\d{2}|\d)", text1)
            rez2 = re.findall("/" + file_ext + " (\d{2}|\d)", text2)
            #print(os.path.join(dirpath, file) + "     " + str(rez0) + "     " + str(rez1) + "     " + str(rez2) )

            
            
            if(rez0): 
                result.append((os.path.join(dirpath, file),rez0))       
            elif(rez1): 
                result.append((os.path.join(dirpath, file),rez1))        
            elif (rez2):
                result.append((os.path.join(dirpath, file),rez2))         
                
                
```

```python
 
y=[]    
                
for i in result:
    y.append(int(i[1][0]))
y
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)
```

```python
print(X)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier( max_iter=1000)

clf.fit(X, y)
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
img1 = cv2.imread('test.jpg')

documents1 = [pytesseract.image_to_string(img1, config=custom_config)]

Y = vectorizer.transform(documents1)
print(Y)
prediction1 = clf.predict_proba(Y)
prediction2 = clf.predict(Y)

print(prediction1)
print(prediction2)
```
