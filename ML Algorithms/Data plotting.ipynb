```python
##imports 

import pandas as pd

from matplotlib import pyplot as plt
```

```python
##Needed CSV's
trainingData = pd.read_csv("train-data.csv", index_col=0) 
testingData  = pd.read_csv("test.csv", index_col=0) 
```

```python
fig, ax = plt.subplots()
trainingData['Label'].plot.kde(ax=ax, legend=False, title='Histogram Labels Training Data')
trainingData['Label'].plot.hist(density=True, ax=ax, legend=False, rwidth=0.9)

plt.show()
```

```python
trainingData.hist('Label',weights=trainingData['Label'], rwidth=0.9)
plt.ylabel('counts')
plt.show()
```

```python
print('Mean: ', trainingData['Label'].mean())
print('Standard deviation: ', trainingData['Label'].std(ddof=1))
print('Variance: ', trainingData['Label'].var(ddof=1))
```

```python
fig, ax = plt.subplots()
testingData['Label'].plot.kde(ax=ax, legend=False, title='Histogram Labels Testing Data')
testingData['Label'].plot.hist(density=True, ax=ax, legend=False, rwidth=0.9)

plt.show()
```

```python
testingData.hist('Label',weights=testingData['Label'], rwidth=0.9 )
plt.ylabel('counts')
plt.show()
```

```python
print('Mean: ', testingData['Label'].mean())
print('Standard deviation: ', testingData['Label'].std(ddof=1))
print('Variance: ', testingData['Label'].var(ddof=1))
```
